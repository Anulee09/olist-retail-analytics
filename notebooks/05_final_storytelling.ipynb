# Olist Retail Analytics: Final Business Story

## Business Problem
An e-commerce company wants to understand what drives revenue, customer retention, and customer satisfaction.

## Objective
To usse SQL and Python to analyze transactional data and recommend actions for growth.

## Data Used
- customers
- orders
- order_items
- products
- payments
- reviews
- sellers
- category_translation

## Key Findings

### 1. Strong Business Growth
Monthly orders increased significantly from 2016 to 2018, indicating rapid platform growth.

### 2. Revenue Concentration
A few product categories generated the majority of revenue, especially Health & Beauty, Watches & Gifts, and Bed/Bath/Table.

### 3. Customer Opportunity
A large share of customers were one-time buyers, suggesting retention opportunities.

### 4. Delivery Drives Satisfaction
On-time deliveries had an average review score of 4.29, while orders delayed 8+ days averaged only 2.41.

## Recommendations

1. Increase marketing investment in top-performing categories.
2. Create loyalty campaigns for one-time customers.
3. Improve delivery reliability and seller monitoring.
4. Use seasonal order trends for inventory planning.
5. Track monthly KPIs in a dashboard.
